# Linear Regression Lab

## Data

```python
import pandas as pd

# TODO Read Training Data CSV
data_train = pd.read_csv("Ice_Cream_Sales_Train.csv")

# Display the first 10 rows of the training data
print(data_train.head())

# Display the shape of the training data
print("\nShape of the training data:",data_train.shape)
```

```python
# TODO Read Testing Data CSV
data_test = pd.read_csv("Ice_Cream_Sales_Test.csv")

# Display the first 10 rows of the test data
print(data_test.head())

# Display the shape of the test data
print("\nShape of the test data:",data_test.shape)
```

```python
# TODO Extract X_train(independent variable) and y_train(dependent variable) from training dataframe
X_train = data_train[['Temperature']]
y_train = data_train['Ice Cream Profits']

# TODO Extract X_test(independent variable) and y_test(dependent variable) from testing dataframe
X_test = data_test[['Temperature']]
y_test = data_test['Ice Cream Profits']
```

```python
import matplotlib.pyplot as plt

# TODO Draw Scatter Plot for Training Data
# hint: check https://www.w3schools.com/python/matplotlib_scatter.asp
plt.scatter(X_train, y_train)


# Add labels and title
plt.xlabel("Temperature")
plt.ylabel("Ice Cream Profits")
plt.title("Scatter Plot of Temperature vs Ice Cream Profits")
```

## Model Training

```python
# TODO Load Linear Regression Model form Sklearn
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
# TODO Fit the model on training data
model.fit(X_train, y_train)
```

**🗒️🤓 Remember Linear Regression simply learns straight line equation**

hint: check https://scikit-learn.org/0.15/modules/generated/sklearn.linear_model.LinearRegression.html in the attributes for the linear regression model to get the coefficient (m) & intercept (b)


```python
# TODO print the coefficient (m) learned by the model
m = model.coef_
print("Coefficient (m) learned by the model:", m)

# TODO print the intercept (b) learned by the model
b = model.intercept_
print("Intercept (b) learned by the model:", b)
```

```python
# Plot the data points and the regression line learned :D
plt.scatter(X_train, y_train, color='blue', label='Training data')
plt.plot(X_train, model.predict(X_train), color='red', label='Regression line')

# Add labels and title
plt.xlabel("Temperature")
plt.ylabel("Ice Cream Profits")
plt.title("Linear Regression on Training Data")

# Add a legend
plt.legend()

# Show the plot
plt.show()
```

## Model Evaluation

```python
# TODO: Predict the model on test data
y_predict = model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
# TODO: Compute Mean Absolute Error (MAE)
mae = mean_absolute_error(y_test, y_predict)

# TODO: Compute Mean Squared Error (MSE)
mse = mean_squared_error(y_test, y_predict)

# Print Mean Squared Error
print(f'Mean Squared Error: {mse}')

# Print Mean Absolute Error
print(f'Mean Absolute Error: {mae}')
```

```python
# Plot the data points and the regression line learned :D
plt.scatter(X_test, y_test, color='blue', label='Testing data')
plt.plot(X_test, model.predict(X_test), color='red', label='Regression line')

# Add labels and title
plt.xlabel("Temperature")
plt.ylabel("Ice Cream Profits")
plt.title("Linear Regression on Testing Data")

# Add a legend
plt.legend()

# Show the plot
plt.show()
```
